**Data Wrangling**

Data wrangling ensures that the dataset is clean, structured, and ready for accurate analysis and visualization. The following transformations are applied to improve data quality and analytical usefulness.

Column Handling

Step 1: Drop Irrelevant or Redundant Columns

Columns like description may not be required for numerical or categorical trend analysis. If the focus is statistical EDA, long text fields can be removed.

```python
# Drop unnecessary column (optional based on analysis goal)
df.drop(columns=['description'], inplace=True)
```

**Justification:**

The description column contains long text data that is not required for trend-based or statistical analysis.

Step 2: Rename Columns for Clarity

```python
# Rename columns for consistency
df.rename(columns={
    'show_id': 'Show_ID',
    'type': 'Content_Type',
    'listed_in': 'Genre',
    'date_added': 'Date_Added',
    'release_year': 'Release_Year'
}, inplace=True)
```

**Justification:**

Renaming improves readability and ensures consistent naming conventions (capitalized, clear meaning).

Step 3: Reorder Columns

```python
# Reorder important columns
df = df[['Show_ID', 'Content_Type', 'title', 'director', 'cast',
         'country', 'Date_Added', 'Release_Year',
         'rating', 'duration', 'Genre']]
```

**Justification:**

Reordering improves logical flow (ID → Content Info → Date → Category).

Feature Creation

Derived features help extract deeper insights from existing columns.

Create Duration in Numeric Format

The duration column contains values like:

"90 min"

"2 Seasons"

We separate numeric duration and unit.

```python
# Extract numeric duration safely
df['Duration_Value'] = df['duration'].str.extract(r'(\d+)').astype(float)

# Extract duration type (min or Seasons)
df['Duration_Type'] = df['duration'].str.extract(r'([a-zA-Z]+)')
```

**Justification:**

Separating duration allows numerical comparison between movies and TV shows.

Create Content Age (Years Since Release)

```python
# Create content age feature
import datetime
current_year = datetime.datetime.now().year

df['Content_Age'] = current_year - df['Release_Year']
```

**Justification:**

Helps analyze how old the content is and study trends of older vs newer releases.

Date & Time Feature Extraction

The dataset contains Date_Added.

Convert to Datetime Format

```python
# Convert Date_Added to datetime
df['Date_Added'] = pd.to_datetime(df['Date_Added'])
```

**Justification:**

Datetime conversion allows extraction of time-based components.

Extract Year, Month, Day, Quarter

```python
df['Added_Year'] = df['Date_Added'].dt.year
df['Added_Month'] = df['Date_Added'].dt.month
df['Added_Day'] = df['Date_Added'].dt.day
df['Added_Quarter'] = df['Date_Added'].dt.quarter
```

**Justification:**

These features help analyze:

Content growth by year

Monthly addition trends

Quarterly content strategy

Data Type Conversion

Ensure columns are stored in appropriate formats.

```python
# Convert categorical columns
categorical_cols = ['Content_Type', 'country', 'rating', 'Genre', 'Duration_Type']
for col in categorical_cols:
    df[col] = df[col].astype('category')

# Convert numeric columns
df['Release_Year'] = df['Release_Year'].astype(int)
df['Duration_Value'] = df['Duration_Value'].astype(float)

# Check updated data types
print(df.dtypes)
```

**Justification:**

Improves memory efficiency

Enhances computational performance

Ensures accurate analysis

Summary

→ Dropped irrelevant column

→ Renamed columns for clarity

→ Reordered columns

→ Created new derived features

→ Converted date column to datetime

→ Extracted year, month, day, quarter

→ Converted data types appropriately

**Data Cleaning and Preparation**

Data cleaning ensures that the dataset is accurate, consistent, and reliable before performing visualization and analysis.

Duplicate Handling

Duplicate records can distort analysis results and must be identified and removed.

Step 1: Identify Duplicate Rows

```python
# Check duplicate rows
duplicate_count = df.duplicated().sum()
print("Number of duplicate rows:", duplicate_count)
```

Step 2: Remove Duplicates

```python
# Remove duplicates
df = df.drop_duplicates()

# Confirm removal
new_duplicate_count = df.duplicated().sum()
print("Duplicates after removal:", new_duplicate_count)
```

Step 3: Report Number of Duplicates Removed

```python
print("Total duplicates removed:", duplicate_count)
```

**Justification:**

Duplicate entries may occur due to data merging or collection errors. Removing duplicates ensures:

Accurate frequency counts

Correct trend analysis

Reliable statistical summaries

Missing Value Treatment

Missing values can affect analysis quality and must be handled appropriately.

Step 1: Identify Missing Values

```python
# Check missing values per column
missing_values = df.isnull().sum()
print("Missing values in each column:\n")
print(missing_values)

# Percentage of missing values
missing_percentage = (df.isnull().sum() / len(df)) * 100
print("\nMissing value percentage:\n")
print(missing_percentage)
```

Step 2: Apply Suitable Treatment

Different columns require different approaches:
Categorical Columns (country, rating, director)

Fill with Mode (most frequent value)

```python
# Fill categorical columns safely
df['country'] = df['country'].fillna(df['country'].mode()[0])
df['rating'] = df['rating'].fillna(df['rating'].mode()[0])
df['director'] = df['director'].fillna("Unknown")

# For cast column
df['cast'] = df['cast'].fillna("Not Available")

# For numerical column
df['Duration_Value'] = df['Duration_Value'].fillna(df['Duration_Value'].median())
```

**Justification:**

Mode maintains category consistency

Replacing director with "Unknown" preserves records without deleting data

Text Columns (cast)

```python
df['cast'] = df['cast'].fillna("Not Available")
```

**Justification:**

Prevents loss of data and keeps dataset complete.

Date Column (Date_Added)

If missing values are small, remove rows:

```python
df = df.dropna(subset=['Date_Added'])
```

**Justification:**

Date is critical for time-based analysis. Removing a small number of rows ensures accurate trend analysis.

Numerical Columns (if any missing)

If numerical columns contain missing values:

```python
df['Duration_Value'] = df['Duration_Value'].fillna(df['Duration_Value'].median())
```

**Justification:**

Median is preferred over mean because:

* It is less affected by outliers

*  Suitable for skewed distributions

Step 3: Final Missing Value Check

```python
print("Remaining missing values:\n", df.isnull().sum())
```

Summary

→ Identified duplicate rows

→ Removed duplicates

→ Reported number removed

→ Identified missing values

→ Used mode for categorical columns

→ Used median for numerical columns

→ Removed critical missing date rows

→ Verified final dataset cleanliness

**Key Insights Summary**

The exploratory data analysis of the Amazon Prime Titles dataset provided comprehensive insights into content distribution, growth trends, audience targeting, and genre concentration. One of the most significant findings is that Movies dominate the platform’s content library compared to TV Shows. Across multiple years, the number of movies added consistently exceeds the number of television series, indicating a strong strategic emphasis on film-based entertainment.

The yearly trend analysis reveals a noticeable increase in content additions during recent years, reflecting aggressive platform expansion and content acquisition strategies. This growth suggests increased competition in the streaming industry and the platform’s effort to strengthen its market presence.

From a rating perspective, the majority of content falls under categories such as 13+, 16+, and TV-14. This indicates that Amazon Prime primarily targets teenage and adult audiences. While family-friendly content rated “ALL” and “7+” is present, it represents a smaller proportion compared to mature and mid-level ratings. This shows clear audience segmentation and strategic content positioning.

Genre analysis reveals that Drama, Comedy, Documentary, and Action are among the most represented categories. These genres tend to attract broad audience engagement and are typically high-performing categories in streaming platforms. In contrast, highly specific or niche genre combinations appear less frequently, indicating focused but limited diversification.

The multivariate analysis further demonstrates that genre and rating are closely related, with certain genres predominantly appearing under specific maturity ratings. This suggests that the platform carefully aligns content themes with appropriate audience classifications.

Overall, the dataset shows structured content expansion, audience-based categorization, and a clear dominance of movies over TV Shows.

Based on the exploratory data analysis, it can be concluded that Amazon Prime Video maintains a diverse yet strategically structured content library.
